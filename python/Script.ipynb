```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def drawFpsPlot(data):
    # Extract columns
    for i in data:
        frames_analyzed = i["Frames Analyzed"]
        elapsed_time = i["Elapsed Time (s)"]
        fps = i["FPS"]
        
        # Calculate the mean FPS
        mean_fps = fps.mean()
        print(f"Mean FPS: {mean_fps}")
        print("")
        
        # Plot the FPS values
        plt.plot(elapsed_time, fps, marker="o", label="FPS")
    
    plt.title("FPS over time")
    plt.xlabel("Elapsed Time (s)")
    plt.ylabel("FPS")
    plt.ylim(bottom=0)
    plt.ylim(top=35)
    plt.legend()
    plt.grid()
    plt.show()
```

```python
data1 = pd.read_csv("desktop_strategy_1.csv")
data2 = pd.read_csv("raspberry_strategy_1.csv")
data = [data1,data2]
drawFpsPlot(data)

print("Rasperby sie zatrzymuje bo java kończy się pamięć (najprowdopodbniej bład w kodzie)")
```
