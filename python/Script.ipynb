```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def drawFpsPlot(data):
    # Extract columns
    frames_analyzed = data["Frames Analyzed"]
    elapsed_time = data["Elapsed Time (s)"]
    fps = data["FPS"]
    
    # Calculate the mean FPS
    mean_fps = fps.mean()
    print(f"Mean FPS: {mean_fps}")
    print("")
    
    # Plot the FPS values
    plt.plot(elapsed_time, fps, marker="o", label="FPS")
    
    plt.title("FPS over time")
    plt.xlabel("Elapsed Time (s)")
    plt.ylabel("FPS")
    plt.ylim(bottom=0)
    plt.ylim(top=35)
    plt.legend()
    plt.grid()
    plt.show()
```

```python
data1 = pd.read_csv("results.csv")
print("Results for desktop computer:")
drawFpsPlot(data1)
```
